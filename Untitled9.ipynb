```python
trainset=pd.read_csv('E:/Download/pre/train.csv')
ad=pd.read_csv('E:/Download/pre/ad.csv')
app_categories=pd.read_csv('E:/Download/pre/app_categories.csv')
user=pd.read_csv('E:/Download/pre/user.csv')
user_app_actions=pd.read_csv('E:/Download/pre/user_app_actions.csv')
user_installedapps=pd.read_csv('E:/Download/pre/user_installedapps.csv')
position=pd.read_csv('E:/Download/pre/position.csv')
temp1=pd.merge(trainset,user,on='userID',how='left')
temp3=pd.merge(temp1,ad,on='creativeID',how='left')
temp4=pd.merge(temp3,app_categories,on='appID',how='left')
temp5=temp4.drop(['clickTime','conversionTime'],axis=1)
final_sample2=temp5.sample(n=30000)
final_label2=final_sample2['label']
final_sample2=final_sample2.drop(['label'],axis=1)
```

```python
cv = StratifiedKFold(final_label3['label'],n_folds=10, shuffle=True, random_state=seed)
```

```python
params_grid = {
    'n_estimators':[300,400,500]
}
params_fixed = {
    'objective': 'binary:logistic',
    'silent': 1,
    'scale_pos_weight':1.0,
    'min_child_weight':1,
    'learning_rate':0.1,
    'gamma':0,
    'max_depth':1
}
```

```python
default_params={
    'objective': 'binary:logistic',
    'silent': 1,
    'min_child_weight':1,
    'learning_rate':0.1,
    'max_depth':3,
    'n_estimators':200,
    'scale_pos_weight':1.0,
    'gamma':0
}
n_estimators_range = np.linspace(1, 200, 10).astype('int')
train_scores, test_scores = validation_curve(
    XGBClassifier(**default_params),
    final_sample3, final_label3,
    param_name = 'n_estimators',
    param_range = n_estimators_range,
    cv=cv,
    scoring='neg_log_loss'
)
```

```python
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)

fig = plt.figure(figsize=(10, 6), dpi=100)

plt.title("Validation Curve with XGBoost (eta = 0.3)")
plt.xlabel("number of trees")
plt.ylabel("log_loss")
plt.ylim(-0.75, 0)

plt.plot(n_estimators_range,train_scores_mean,label="Training score",color="r")

plt.plot(n_estimators_range,test_scores_mean,label="Cross-validation score",color="g")

plt.fill_between(n_estimators_range,train_scores_mean - train_scores_std,train_scores_mean + train_scores_std,alpha=0.2, color="r")

plt.fill_between(n_estimators_range,test_scores_mean - test_scores_std,test_scores_mean + test_scores_std,alpha=0.2, color="g")

plt.axhline(y=1, color='k', ls='dashed')

plt.legend(loc="best")
plt.show()

i = np.argmax(test_scores_mean)
print("Best cross-validation result ({0:.4f}) obtained for {1} trees".format(test_scores_mean[i], n_estimators_range[i]))
```

```python
bst_grid = GridSearchCV(
    estimator=XGBClassifier(**params_fixed, seed=seed),
    param_grid=params_grid,
    cv=cv,
    scoring='neg_log_loss'
)
```

```python
bst_grid.fit(final_sample3, final_label3)
for key, value in bst_grid.best_params_.items():
    print("\t{}: {}".format(key, value))
```
